# Ask the stock name and display Quarterly Results of the stock name

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Ask the user for the stock name
stock_name = input("Enter the stock name: ")

# Define the URL of the company's consolidated page
url = f'https://www.screener.in/company/{stock_name}/consolidated/'

# Send a request to fetch the HTML content
response = requests.get(url)
if response.status_code != 200:
    raise Exception(f"Failed to load page {url}")

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the table containing the quarterly results
table = soup.find('table', {'class': 'data-table responsive-text-nowrap'})

# Check if the table is found
if not table:
    raise Exception("Failed to find the quarterly results table.")

# Extract the table headers if they exist
headers = [th.text.strip() for th in table.find('thead').find_all('th')] if table.find('thead') else []

# Extract the table rows
rows = []
tbody = table.find('tbody')
if tbody:
    for tr in tbody.find_all('tr'):
        cells = [td.text.strip() for td in tr.find_all('td')]
        rows.append(cells)

# Check if headers and rows are populated
if not headers or not rows:
    raise Exception("Failed to extract data from the table.")

# Create a DataFrame from the extracted data
df = pd.DataFrame(rows, columns=headers)

# Print the DataFrame
print(df)

# Save the DataFrame to a CSV file with the stock name included in the file name
# csv_file_name = f'quarterly_results_{stock_name}.csv'
# df.to_csv(csv_file_name, index=False)
# print(f"Data saved to {csv_file_name}")
```
